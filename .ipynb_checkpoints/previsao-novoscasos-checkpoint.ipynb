```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
colunas_usadas = [tf.feature_column.numeric_column(key=c) for c in colunas]
```

```python
print(colunas_usadas[0])
```

```python
from sklearn.model_selection import train_test_split
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(x,y,test_size=0.5)
```

```python
funcao_treinamento = tf.estimator.inputs.pandas_input_fn(x = x_treinamento, y = y_treinamento,batch_size=32,num_epochs=None,shuffle=True)
```

```python
funcao_teste = tf.estimator.inputs.pandas_input_fn(x = x_teste, y = y_teste, batch_size=32,num_epochs=10000, shuffle=False)
```

```python
regressor = tf.estimator.LinearRegressor(feature_columns=colunas_usadas)
```

```python
regressor.train(input_fn=funcao_treinamento,steps=10000)
```

```python
metricar_treinameto = regressor.evaluate(input_fn=funcao_treinamento, steps=10000)
```

```python
metricar_teste = regressor.evaluate(input_fn=funcao_teste, steps=10000)
```
